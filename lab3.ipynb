## Lab 3 - Bài 1

```python
# Code lab 3 bài 1 ở đây
tong = 0
dem = 0
day_so_chia_het_cho_3 = []
day_so = []
while True:
    so = input("Nhập số: ")
    if so == 'q':
        print("Thoat chương trình")
        break
    else:
        try:
            phan_tu = int(so)
            day_so.append(phan_tu)
        except ValueError:
            print("Vui lòng nhập đúng số hoặc q để thoát chương trình")
            continue
print("Dãy số vừa nhập là: ", day_so)
# Sắp xếp dãy số tăng dần
day_so.sort()
print("Dãy số sau khi sắp xếp tăng dần là: ", day_so)
print("giá trị lớn nhất trong dãy số là: ", (day_so[-1]))
print("giá trị nhỏ nhất trong dãy số là: ", (day_so[0]))
# Sắp xếp dãy số giảm dần
day_so.sort(reverse=True)
print("Dãy số sau khi sắp xếp giảm dần là: ", day_so)
print("giá trị lớn nhất trong dãy số là: ", (day_so[0]))
print("giá trị nhỏ nhất trong dãy số là: ", (day_so[-1]))

#duyện danh sách phần tử 

for phan_tu in day_so:
    if phan_tu % 3 == 0:
        tong += phan_tu
        dem += 1
        day_so_chia_het_cho_3.append(phan_tu)
print("Tổng các phần tử chia hết cho 3 là: ", tong)
print(f"có tất cả {dem} phần tử chia hết cho 3")
if dem != 0:
    print("Trung bình cộng các phần tử chia hết cho 3 là: ", tong/dem)
else:
    print("Không có phần tử nào chia hết cho 3")    
print("Các phần tử chia hết cho 3 là: ", day_so_chia_het_cho_3)
```

## Lab 3 - Bài 2

```python
# Code lab 3  bài 2 ở đây
# Viết chương trình nhập dữ liệu của một sinh viên bao gồm: họ tên, năm sinh, điểm trung bình.

danh_sach_sv = [] 

so_luong = int(input("Nhập số lượng sinh viên: "))

for i in range(so_luong):
    print(f"\n--- Nhập thông tin sinh viên {i+1} ---")
    ho_ten = input("Nhập họ tên: ")
    nam_sinh = int(input("Nhập năm sinh: "))
    diem_tb = float(input("Nhập điểm trung bình: "))


    # Tạo dictionary cho từng sinh viên
    sinh_vien = {
        "Họ tên": ho_ten,
        "Năm sinh": nam_sinh,
        "Điểm trung bình": diem_tb
    }



    # Xét học lực
    if diem_tb >= 8:
        sinh_vien["Học lực"] = "Giỏi"
    elif diem_tb >= 6.5:
        sinh_vien["Học lực"] = "Khá"
    elif diem_tb >= 5:
        sinh_vien["Học lực"] = "Trung bình"
    else:
        sinh_vien["Học lực"] = "Yếu"


    # Thêm sinh viên vào danh sách
    danh_sach_sv.append(sinh_vien)


# Xuất thông tin tất cả sinh viên
print("\n===== DANH SÁCH SINH VIÊN =====")
for idx, sv in enumerate(danh_sach_sv, start=1):
    print(f"\nSinh viên {idx}:")
    for key, value in sv.items():
        print(f"{key}: {value}")
```

## Lab 3 - Bài 3

```python
# Code lab 3  bài 3 ở đây
# Viết chương trình nhập một số nguyên từ bàn phím và cho biết số đó có phải số nguyên tố hay không.
n = int(input("Nhập một số nguyên: "))

ok = True
i = 2

while i < n:   
    if n % i == 0: 
        ok = False
        break
    i += 1

if n <= 1:
    print(f"{n} không phải là số nguyên tố.")
elif ok:
    print(f"{n} là số nguyên tố.")
else:
    print(f"{n} không phải là số nguyên tố.")
```

## Lab 3 - Bài 4

```python
# code lab 3 bài 4 ở đây

MAX_I = 20   # số cột: bảng 1..9
MAX_J = 10   # số dòng: nhân 1..9

def cell(i, j):
    return f"{i} x {j} = {i*j}"

# Tính độ rộng ô theo ô dài nhất ("9 x 9 = 81")
cell_width = len(cell(MAX_I, MAX_J))  # thường là 10

# In lưới: mỗi dòng là j = 1..9, mỗi cột là i = 1..9
for j in range(1, MAX_J + 1):
    row = []
    for i in range(1, MAX_I + 1):
        row.append(f"{cell(i, j):<{cell_width}}")
    print("  ".join(row))  
```
