```python
from DanhSachNhanVien import *
danh_sach_nhan_vien = []


while True:
    print("""
    ================== CHƯƠNG TRÌNH QUẢN LÝ NHÂN VIÊN ==================
    1. Nhập danh sách nhân viên từ bàn phím và lưu vào file.
    2. Đọc thông tin nhân viên từ file và xuất ra màn hình.
    3. Tìm và hiển thị nhân viên theo mã.
    4. Xóa nhân viên theo mã.
    5. Cập nhật thông tin nhân viên theo mã.
    6. Tìm các nhân viên theo khoảng lương.
    7. Sắp xếp nhân viên theo họ và tên.
    8. Sắp xếp nhân viên theo thu nhập.
    9. Xuất 5 nhân viên có thu nhập cao nhất.
    0. Thoát chương trình.
    ====================================================================
    """)

    lua_chon = input("Mời bạn chọn chức năng: ")

    match lua_chon:
        case "1":
            print("Đã chọn chức năng nhập nhân viên và lưu file.")
            print("\nChức năng 1: Nhập danh sách nhân viên")
            dsnv = DanhSachNhanVien()
            dsnv.nhap_va_luu()
            print("-"*40)   


        case "2":
            print("Đã chọn chức năng đọc file và xuất danh sách.")
            print("\nChức năng 2: Đọc và xuất danh sách nhân viên")
            with open("nhanvien.txt", "r", encoding="utf-8") as file:
                noidung = file.read()
                print(noidung)
                print("-"*40)
            

        case "3":
            print("Đã chọn chức năng tìm kiếm nhân viên.")
            print("\nChức năng 3: Tìm nhân viên theo mã")
            dsnv = DanhSachNhanVien()
            dsnv.tim_theo_ma()
            print("-"*40)  


        case "4":
            print("Đã chọn chức năng xóa nhân viên.")
            print("\nChức năng 4: Xóa nhân viên theo mã")
            dsnv = DanhSachNhanVien()
            dsnv.xoa_theo_ma()
            print("-"*40)
        
            

        case "5":
            print("Đã chọn chức năng cập nhật thông tin.")
            print("\nChức năng 5: Cập nhật thông tin nhân viên")
            dsnv = DanhSachNhanVien()
            dsnv.cap_nhat_theo_ma()
            print("-"*40)
            
            

        case "6":
            print("Đã chọn chức năng tìm theo khoảng lương.")
            print("\nChức năng 6: Tìm nhân viên theo khoảng lương nhập từ bàn phím")
            dsnv = DanhSachNhanVien()
            dsnv.tim_theo_khoang_luong()
            print("-"*40)

            

        case "7":
            print("Đã chọn chức năng sắp xếp theo tên.")
            print("\nChức năng 7: Sắp xếp nhân viên theo họ và tên")
            dsnv = DanhSachNhanVien()
            dsnv.sap_xep_theo_ho_ten()
            print("-"*40)

            

        case "8":
            print("Đã chọn chức năng sắp xếp theo thu nhập.")
            print("\nChức năng 8: Sắp xếp nhân viên theo thu nhập")
            dsnv = DanhSachNhanVien()
            dsnv.sap_xep_theo_thu_nhap()
            print("-"*40)

            

        case "9":
            print("Đã chọn chức năng xuất 5 nhân viên lương cao nhất.")
            print("\nChức năng 9: Xuất 5 nhân viên có thu nhập cao nhất")
            dsnv = DanhSachNhanVien()
            dsnv.xuat_5_nhan_vien_luong_cao_nhat()
            print("-"*40)
            
            

        case "0":
            print("\nCảm ơn bạn đã sử dụng chương trình. Tạm biệt!")
            break

        case _:
            print("\nLựa chọn không hợp lệ. Vui lòng chọn lại từ 0 đến 9.")

    # Thêm một input để người dùng có thể xem kết quả trước khi menu hiện lại
    input("\nNhấn Enter để tiếp tục...")
```
